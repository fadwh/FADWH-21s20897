```python
# Define paths to the dataset
train_dir = '/content/drive/MyDrive/train'  # Update with your actual path
test_dir = '/content/drive/MyDrive/test'    # Update with your actual path
```

```python
import os
print(len(os.listdir('/content/drive/MyDrive/train/rugby')))
print(len(os.listdir('/content/drive/MyDrive/train/soccer')))
print(len(os.listdir('/content/drive/MyDrive/test/rugby')))
print(len(os.listdir('/content/drive/MyDrive/test/soccer')))
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Create ImageDataGenerator for training set
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Split 20% of the images for validation
)

# Load and prepare training data
train_data = train_datagen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',  # 'binary' for binary classification (cats vs. dogs)
    subset='training'  # Specify 'training' for the training set
)

# Create ImageDataGenerator for validation set
validation_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # Note: Using the same validation split as in the training set
)

# Load and prepare validation data
validation_data = validation_datagen.flow_from_directory(
    train_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary',
    subset='validation'  # Specify 'validation' for the validation set
)
```

```python
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten

custom_model = Sequential()    #ResNet50 is a pre-trained model
base_model = tf.keras.applications.ResNet50(include_top = False, #because i have my own image of a diff dimension
                                                input_shape = (224,224,3),
                                                pooling = 'max', classes = 2,
                                                weights = 'imagenet')
for layer in base_model.layers:
    layer.trainable = False       #all layers in the Resnet50 is not trainable; keep those weights; I am going to add my own i/p and other layers

custom_model.add(base_model)
custom_model.add(Flatten())
custom_model.add(Dense(128, activation = 'relu'))
custom_model.add(Dense(64, activation = 'sigmoid'))
```

```python
custom_model.summary()
```

```python
from keras.optimizers import Adam
custom_model.compile(optimizer=Adam(learning_rate=0.001),loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history = custom_model.fit(train_data, epochs=5, validation_data=validation_data)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'],color='red',label='train')
plt.plot(history.history['val_accuracy'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
plt.plot(history.history['loss'],color='red',label='train')
plt.plot(history.history['val_loss'],color='blue',label='validation')
plt.legend()
plt.show()
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224,224),
    batch_size=32,
    class_mode='binary'
)
```

```python
#predict the test data
predictions = custom_model.predict(test_data)
print(predictions)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Assuming `validation_data` is used for evaluation
# Get the ground truth labels
true_labels = validation_data.classes

# Predict using the custom CNN model
custom_preds = custom_model.predict(validation_data)
custom_preds = custom_preds.argmax(axis=1)

# Predict using the pre-trained ResNet50 model
pretrained_preds = base_model.predict(validation_data)
pretrained_preds = pretrained_preds.argmax(axis=1)

# Calculate metrics for custom CNN model
custom_accuracy = accuracy_score(true_labels, custom_preds)
custom_f1 = f1_score(true_labels, custom_preds, average='weighted')

# Calculate metrics for pre-trained ResNet50 model
pretrained_accuracy = accuracy_score(true_labels, pretrained_preds)
pretrained_f1 = f1_score(true_labels, pretrained_preds, average='weighted')

# Print classification report
print("Custom CNN Model Classification Report:")
print(classification_report(true_labels, custom_preds))

print("Pre-trained ResNet50 Model Classification Report:")
print(classification_report(true_labels, pretrained_preds))

# Compile results into a table
import pandas as pd

results = {
    'Model': ['Custom CNN', 'Pre-trained ResNet50'],
    'Accuracy': [custom_accuracy, pretrained_accuracy],
    'F1-Score': [custom_f1, pretrained_f1]
}

results_df = pd.DataFrame(results)
print(results_df)
```
